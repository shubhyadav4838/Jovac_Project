```python

# Machine Learning Online Assignment : NumPy & Pandas

# PART I - NUMPY

# Task 1: Creating Arrays
# ------------------------
import numpy as np

# 1.1 - Creating a 1D array of integers from 1 to 10
array_1d = np.arange(1, 11)

# 1.2 - Creating a 2D array (3x3) from 1 to 9
array_2d = np.arange(1, 10).reshape(3, 3)

# 1.3 - Creating a 3D array of shape (3, 5, 3) with random floats
array_3d = np.random.rand(3, 5, 3)

# 2 - Display the arrays
print("1D Array:\n", array_1d)
print("\n2D Array:\n", array_2d)
print("\n3D Array:\n", array_3d)

# Display the shape of each array
print("\nShapes:")
print("1D array shape:", array_1d.shape)
print("2D array shape:", array_2d.shape)
print("3D array shape:", array_3d.shape)

# Display the size (total number of elements) of each array
print("\nSizes:")
print("1D array size:", array_1d.size)
print("2D array size:", array_2d.size)
print("3D array size:", array_3d.size)

# Display the data type of each array
print("\nData Types:")
print("1D array dtype:", array_1d.dtype)
print("2D array dtype:", array_2d.dtype)
print("3D array dtype:", array_3d.dtype)

# Task 2:  Array Indexing and Slicing
# ----------------------------
# 1. Creating the NumPy array from list
data = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])

# 2. Indexing and slicing operations
print("Original array:", data)

# Retrieve the first three elements
print("First three elements:", data[:3])

# Retrieve every alternate element
print("Every alternate element:", data[::2])

# Reverse the array
print("Reversed array:", data[::-1])


# Task 3: Mathematical Operations
# -----------------------
# 1. Create arrays A and B
A = np.random.randint(1, 21, 5)
B = np.random.randint(1, 21, 5)

print("Array A:", A)
print("Array B:", B)

# 2.1 Add, subtract, multiply, and divide the arrays element-wise
print("Element-wise Addition:", A + B)
print("Element-wise Subtraction:", A - B)
print("Element-wise Multiplication:", A * B)
print("Element-wise Division:", A / B)

# 2.2 Compute the dot product of the arrays
print("Dot Product of A and B:", np.dot(A, B))

# 2.3 Mean, median, standard deviation, and variance of array A
print("Mean of A:", np.mean(A))
print("Median of A:", np.median(A))
print("Standard Deviation of A:", np.std(A))
print("Variance of A:", np.var(A))

# 2.4 Maximum and minimum values in array B and their indices
print("Maximum value in B:", np.max(B))
print("Index of Maximum value in B:", np.argmax(B))
print("Minimum value in B:", np.min(B))
print("Index of Minimum value in B:", np.argmin(B))


# Task 4:  Reshaping and Transposing 
# -----------------
# Create a 1D NumPy array of 12 integers ranging from 1 to 12
arr = np.arange(1, 13)
print("Original 1D array:", arr)

# Reshape the array into a 2D array of shape (4, 3)
print("Reshaped into 2D array (4, 3):\n", arr.reshape(4, 3))

# Reshape the array into a 3D array of shape (2, 2, 3)
print("Reshaped into 3D array (2, 2, 3):\n", arr.reshape(2, 2, 3))

# Transpose the reshaped 2D array and display its shape
print("Shape of Transposed 2D array (4, 3) -> (3, 4):", arr.reshape(4, 3).T.shape)

# Task 5:  Boolean Masking and Filtering 
# ------------------------
# Create a NumPy array with 15 random integers between 10 and 50
arr = np.random.randint(10, 51, 15)
print("Original array:", arr)

# Find all elements greater than 25
print("Elements greater than 25:", arr[arr > 25])

# Replace all elements less than 30 with 0
arr[arr < 30] = 0
print("Array after replacing elements less than 30 with 0:", arr)

# Count the number of elements divisible by 5
print("Number of elements divisible by 5:", np.sum(arr % 5 == 0))



# Task 6:  Working with Built-in Functions
# ---------------------------

# Create an array of 10 equally spaced values between 0 and 1
print("Equally spaced values between 0 and 1:", np.linspace(0, 1, 10))

# Create an identity matrix of size 4x4
print("Identity matrix of size 4x4:\n", np.eye(4))

# Generate a 1D array of 20 random integers between 1 and 100
arr = np.random.randint(1, 101, 20)
print("Original random array:", arr)

# Sort the array and find the 5 largest elements
print("5 largest elements:", np.sort(arr)[-5:])


# Task 7: Generic
# ----------------
import time

# Generate two large random arrays of size (100, 100)
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)

# Measure time taken for matrix multiplication
start = time.time()
C = np.dot(A, B)
end = time.time()
print("Time taken for matrix multiplication:", end - start, "seconds")

# Try computing the determinant and inverse of the resulting matrix
try:
    print("Determinant of result matrix:", np.linalg.det(C))
    print("Inverse of result matrix:\n", np.linalg.inv(C))
except np.linalg.LinAlgError:
    print("Inverse not possible: Matrix is singular or not invertible.")

```

```python
# PART II - PANDAS

import pandas as pd

# Task 1: Working with Series
# --------------
import pandas as pd

# Task 1: Working with Series
# ---------------------------

# 1. Create a Pandas Series from the following list
#    data = [25, 30, 35, 40, 45]
#    and assign custom indices ['A', 'B', 'C', 'D', 'E']
s = pd.Series([25, 30, 35, 40, 45], index=['A', 'B', 'C', 'D', 'E'])

# 2. Display the first three elements of the Series
print("First three elements:\n", s[:3])

# 3. Calculate the mean, median, and standard deviation of the Series
print("Mean:", s.mean())
print("Median:", s.median())
print("Standard Deviation:", s.std())


# Task 2: Creating and Inspecting DataFrames
# -----------------
# 1. Create a Pandas DataFrame with the given data
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
})

# 2. Display the first two rows of the DataFrame
print("First two rows of the DataFrame:\n", df.head(2))

# 3. Display the column names, data types, and summary statistics
print("\nColumn names:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nSummary statistics:\n", df.describe(include='all'))

# 4. Add a new column 'Passed' where value is True if Marks >= 80, else False
df['Passed'] = df['Marks'] >= 80
print("\nDataFrame with 'Passed' column:\n", df)

# Task 3:  Data Selection and Filtering 
# --------------------------
# Select and display the Name and Marks columns
print("Selected columns (Name and Marks):\n", df[['Name', 'Marks']])

# Filter and display records where Marks > 80
print("\nStudents with Marks > 80:\n", df[df['Marks'] > 80])

# Display the record of the student with the highest marks
print("\nStudent(s) with the highest marks:\n", df[df['Marks'] == df['Marks'].max()])

# Task 4: Handling Missing Data
# ---------------------
# Step 1: Introduce missing values
df.loc[1, 'Marks'] = None   # Set Marks of 2nd student (index 1) to None
df.loc[4, 'Age'] = None     # Set Age of 5th student (index 4) to None

# Step 2: Identify missing values
print("Missing values in the DataFrame:\n", df.isnull())

# Step 3: Fill missing values in 'Marks' column with the mean of the column
df['Marks'].fillna(df['Marks'].mean(), inplace=True)

# Step 4: Drop rows where the 'Age' column has missing values
df.dropna(subset=['Age'], inplace=True)

# Display the updated DataFrame
print("\nUpdated DataFrame after handling missing data:\n", df)

# Task 5:Grouping and Aggregation
# ----------------
# Group the data by 'Gender' and calculate the mean of 'Age' and 'Marks' for each group
print("Mean Age and Marks by Gender:")
print(df.groupby('Gender')[['Age', 'Marks']].mean())

# Count the number of students in each gender group
print("\nNumber of students by Gender:")
print(df.groupby('Gender').size())

# Task 6: Reading and Writing Data (CSV)
# -----------
# Save the modified DataFrame to a CSV file named "students_data.csv"
df.to_csv("students_data.csv", index=False)

# Read the CSV file into a new DataFrame
df2 = pd.read_csv("students_data.csv")

# Display the first five rows of the newly loaded DataFrame
print("First five rows of the loaded DataFrame:")
print(df2.head())

# Task 7:  General {EDA (using seaborn dataset)}
# -----------------------------------
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset from Seaborn
tips = sns.load_dataset("tips")

# Display the first few rows of the dataset
print("First 5 records of the dataset:")
print(tips.head())

# Display basic statistics (count, mean, std, min, max, etc.)
print("\nDescriptive statistics:")
print(tips.describe())

# Check for missing values in the dataset
print("\nMissing values in each column:")
print(tips.isnull().sum())

# ------------------------
# Data Visualization
# ------------------------

# Boxplot: Distribution of total bill amounts for each day of the week
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()

# Histogram: Distribution of tip amounts
plt.hist(tips['tip'], bins=10, edgecolor='black')
plt.title("Distribution of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")
plt.show()

# Scatterplot: Total bill vs Tip
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Tip vs Total Bill (Colored by Gender)")
plt.show()

# ------------------------
# Summary (Markdown-style comments)
# ------------------------

# - The 'tips' dataset contains 244 records with no missing values.
# - The average total bill is around $19.79 with a standard deviation of ~8.90.
# - Tips range mostly between $2 and $4.
# - Boxplot shows higher total bills on weekends (Sat & Sun).
# - Scatterplot indicates a positive correlation between total bill and tip, and slight gender-based variation.

```
